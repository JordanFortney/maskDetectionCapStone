```python
import numpy as np
import pandas as pd
import tensorflow as tf
from  matplotlib import pyplot as plt

import glob
import cv2
```

```python
yesData = []
yesFiles = glob.glob ("C:/Users/fortn/Documents/BrainStation/CapStone/Project/mask_detection/yes/*.jpg")
for fileName in yesFiles:
    image = cv2.imread(fileName)
    yesData.append(image)
                      
print(f"yesData shape = {np.array(yesData, dtype=object).shape}")
```

```python
noData = []
yesFiles = glob.glob ("C:/Users/fortn/Documents/BrainStation/CapStone/Project/mask_detection/no/*.jpg")
for fileName in yesFiles:
    image = cv2.imread(fileName)
    noData.append(image)
                      
print(f"yesData shape = {np.array(noData, dtype=object).shape}")
```

```python
print(yesData[0].shape)
print(noData[1].shape)
print(yesData[2].shape)
print(noData[3].shape)
print(yesData[4].shape)
```

```python
len(yesData)
```

```python
lenYes = np.arange(0,len(yesData))
lenNo = np.arange(0,len(noData))
height = []

for h in lenYes:
    height.append(yesData[h][0].shape)

for h in lenNo:
    height.append(noData[h][0].shape)
    
print(len(height))
```

```python
width = []

for h in lenYes:
    width.append(yesData[h][1].shape)

for h in lenNo:
    width.append(noData[h][1].shape)
    
print(len(width))
```

```python
print(f"The smallest height is : {min(height)}")
print(f"The smallest height is : {min(width)}")
```

```python
minHeight = min(height)[0]
minWidth = min(width)[0]
minDim = (minWidth, minHeight)
```

```python
for d in lenYes:
    yesData[d] = cv2.resize(yesData[d], minDim)
for d in lenNo:
    noData[d] = cv2.resize(noData[d], minDim)
```
